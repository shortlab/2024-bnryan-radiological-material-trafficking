```python
import csv
import matplotlib.pyplot as plt
direct = '/Users/briannaryan/Desktop/CEvNS/Single Isotope Neutrino Generator/tests/'
```

```python
# Read Spectrum CSV File
def read_csv(iso) :
    spec_file = direct + iso + '/' + iso + '_Neutrino_Spectrum.csv'
    energies = []
    counts = []

    with open(spec_file) as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader :
            energies.append(float(row[0])*10**3)
            counts.append(float(row[1]))
            
    return energies, counts
```

```python
energies, counts = read_csv('Cs-137')
plt.plot(energies, counts, color='black')
plt.xlabel('Neutrino Energy (keV)', fontsize=13)
plt.ylabel('dN/dE (Neutrinos/keV/Decay)', fontsize=13)
#plt.yscale('log')
plt.savefig('Cs-137_paper_spectrum.png')
```
